```python
# creating an image
import numpy as np
import cv2
import matplotlib.pyplot as plt


# store image path
img_path = "./dolphin.jpg"
# load image as gray-scaled
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# blur image before-hand
img = cv2.GaussianBlur(img, (5, 5), 0)

# compute Canny edges
I_edge = cv2.Canny(img, 10, 100)
#I_edge = cv2.Canny(img, 50, 100)

# display images
plt.figure(figsize=(10, 10))
plt.subplot(111), plt.imshow(I_edge, cmap="gray")
plt.title("I_edge"), plt.xticks([]), plt.yticks([])
plt.show()
```
